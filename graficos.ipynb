```python
import matplotlib.pyplot as plt

# Dados fictícios (substitua pelos dados que você obteve)
tamanhos = [100, 1000, 10000, 50000, 100000]

tempos_selection_sort = [0.000000, 0.000000, 0.139000, 3.443000, 13.236000]
tempos_insertion_sort = [0.000000, 0.000000, 0.117000, 3.024000, 9.688000 ]
tempos_bubble_sort =    [0.000000, 0.000000, 0.369000, 8.669000, 37.402000]
tempos_merge_sort =     [0.000000, 0.000000, 0.000000, 0.019000, 0.029000 ]
tempos_quick_sort =     [0.000000, 0.000000, 0.000000, 0.002000, 0.018000 ]
tempos_heap_sort =      [0.000000, 0.000000, 0.016000, 0.030000, 0.032000 ]
```

```python
# Configurando o gráfico
plt.figure(figsize=(10, 5))

plt.plot(tamanhos, tempos_selection_sort, label="Selection Sort", marker="o")
plt.plot(tamanhos, tempos_insertion_sort, label="Insertion Sort", marker="o")
plt.plot(tamanhos, tempos_bubble_sort, label="Bubble Sort", marker="o")
plt.plot(tamanhos, tempos_merge_sort, label="Merge Sort", marker="o")
plt.plot(tamanhos, tempos_quick_sort, label="Quick Sort", marker="o")
plt.plot(tamanhos, tempos_heap_sort, label="Heap Sort", marker="o")


plt.xlabel("Tamanho do Array")
plt.ylabel("Tempo de Execução (segundos)")
plt.title("Desempenho dos Algoritmos de Ordenação")
plt.legend()

# Define o eixo x como logarítmico e ajusta a rotação dos rótulos
#plt.xscale("log")
plt.xticks([100, 1000, 10000, 50000, 100000], rotation=60)
plt.yticks([0, 3, 6, 9, 12, 15, 18, 21, 24, 27, 30, 33, 36, 39])

# Define o intervalo dos eixos
plt.ylim(0, 40)

# Grade e layout
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()

plt.show()
```
